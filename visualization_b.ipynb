```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Read CSV File using pandas

### This is table for ambient tempretur during may and Jun

```python
df=pd.read_csv('./csv/Plant_1_Weather_Sensor_Data.csv')
df.head()
```

### Convert DATE_TIME column to " time data type "

```python
df['Months']=pd.to_datetime(df.DATE_TIME)


```

### Split the Days about from date and create a new column for Day

```python
df['Day']=df.Months.dt.day
df.head()
```

```python
### Group the days and sum its tempretue
```

```python
# df['Month']
df_weather=df.groupby('Day',as_index=False).sum()
df_weather.head(30)
```

## Run seaborn

```python
sns.set()
```

```python
## Plt the seaborn
```

```python
sns.barplot(x='Day',y='AMBIENT_TEMPERATURE',data=df_weather)
```
